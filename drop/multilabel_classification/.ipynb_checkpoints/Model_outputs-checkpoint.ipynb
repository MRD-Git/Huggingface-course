```python
# https://stackoverflow.com/questions/62327803/having-6-labels-instead-of-2-in-hugging-face-bertforsequenceclassification
# 
from transformers import BertForSequenceClassification, BertConfig, BertTokenizer
import torch

config = BertConfig.from_pretrained('bert-base-cased')
config.num_labels = 3
model = BertForSequenceClassification(config) 
print(model.parameters)
config
```

```python
#from transformers import BertTokenizer, BertForSequenceClassification
#import torch
#from mymodel import MultiLabelBert
tokenizer = BertTokenizer.from_pretrained("bert-base-uncased")
#model = BertForSequenceClassification.from_pretrained("bert-base-uncased", num_labels=3)
# inputs
inputs = tokenizer(
    ["Hello, my dog is cute", "Good morning"],
    padding=True,
    batched=True,
    truncation=True,
    return_tensors="pt"
)
print(f"inputs:\n{inputs}")
print(f"\ninputs size:\n{inputs['input_ids'].size(0)}")
# labels
labels = torch.tensor([0, 2])
print(f"\nlabels:\n{labels}")
print(f"\nlabels size:\n{labels.size(0)}")
# outputs
outputs = model(**inputs, labels=labels)
#print(f"\noutputs:\nloss:\t{outputs.loss}\nlogits:\t{outputs.logits}")
outputs
```
